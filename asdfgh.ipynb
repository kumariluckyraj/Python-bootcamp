```python
import matplotlib.pyplot as plt


import numpy as np
x=np.arange(0,10)
y=np.arange(11,21)
plt.plot(x,y,'g--',linewidth=9,markersize=12)
plt.xlabel('hehe')
plt.ylabel('haha')
plt.title('test')
plt.savefig('Test.png')
plt.show()
```

```python
#subplots
plt.subplot(4,2,1)#nrow, ncol , index
plt.plot(x,y,'r--')
plt.show()
```

```python
labels='a','b','c','d'
sizes= [123,34,5,466]
colors=['yellow','red','orange','green']
explode=(0.4,0,0.2,0) #how much far u want that perticular slice to be
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%', shadow=False)
plt.axis('equal')
plt.show
```
